```python
import pandas as pd
url = 'https://raw.githubusercontent.com/sabah0312/ml-datasets/refs/heads/main/regression_toy_data'
df = pd.read_csv(url,   delim_whitespace=True)

df
```

```python
import torch

X = torch.tensor(df['X'].values, dtype= torch.float32)
y = torch.tensor(df['y'].values, dtype= torch.float32)
```

```python
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
class Perceptron:
  def __init__(self, num_features, learning_rate = 0.01):
    self.w = torch.randn(num_features, 1, requires_grad=True)
    self.bias = torch.randn(1, requires_grad=True)
    self.learning_rate = learning_rate

  def forward(self, x):
    return  x @ self.w + self.bias

  def update(self, x, y_true):
    prediction = self.forward(x)
    error = (y_true - prediction)**2
    error = error.mean()

    error.backward()
    # print(f"weight grads: {self.w}")
    # print(f"bias grads: {self.bias}")
    with torch.no_grad():
      self.bias -= self.learning_rate * self.bias.grad
      self.w -= self.learning_rate * self.w.grad

    self.w.grad.zero_()
    self.bias.grad.zero_()
    return error
```

```python
PPn = Perceptron(1)
PPn.update(X[1], y[1])
```

```python
PPn = Perceptron(1)

print(f"initial weights: {PPn.w}")
print(f"initial bias : {PPn.bias}")
epoch_no = 150
for epoch in range(epoch_no):
  error_total = 0
  for i in range(len(X)):
    x = X[i]
    error = PPn.update(x, y[i])
    error_total += abs(error)
  if epoch%10 == 0:
    print(f"epoch: {epoch}/ {epoch_no}, error: {error_total}")
print(f" weights: {PPn.w}")
print(f" bias : {PPn.bias}")
```
